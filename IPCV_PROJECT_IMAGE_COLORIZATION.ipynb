```python
from google.colab import drive
import cv2
from google.colab.patches import cv2_imshow
```

```python
drive.mount('/content/drive')
```

```python
sample_image = cv2.imread("/content/drive/MyDrive/ColorfulColorization/Apple/Apple1.jpg")
cv2_imshow(sample_image)
```

```python
lab_image = cv2.cvtColor(sample_image,cv2.COLOR_RGB2Lab)

#extracting individual components
L,a,b = cv2.split(lab_image)

#displaying images in each channel
cv2_imshow(L) #image in L channel
cv2_imshow(a) #image in a channel
cv2_imshow(b) #image in b channel
```
